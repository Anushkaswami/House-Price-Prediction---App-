```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score 
from xgboost import XGBRegressor 
import joblib
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
m_values = missing_values[missing_values > 500]
plt.figure(figsize=(10,8))
sns.barplot(x=m_values.index, y=m_values.values, palette='viridis')
plt.title('Missing Values By Feature')
plt.xlabel("Count of missing values")
plt.ylabel("Features")
plt.show()
```

```python
df = df.drop(columns=['Id','PoolQC','FireplaceQu','Fence','Alley','MiscFeature','MasVnrType'])
df
```

```python
df.columns
```

```python
c_columns = df.select_dtypes(include=['object','category']).columns
print(c_columns)
```

```python
label_encoder = LabelEncoder()
c_columns = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',
       'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
       'SaleType', 'SaleCondition']
# Use an instance of LabelEncoder for each column and handle missing values robustly
for col in c_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].fillna('missing').astype(str))
df
```

```python
corr_matrix = df.corr()
corr_with_price = corr_matrix['SalePrice'].sort_values(ascending=False)
print("Correlated Features", corr_with_price.head(20))
```

```python
c_features = corr_with_price[corr_with_price>0.25].index
df = df[c_features]
df
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[corr_with_price[corr_with_price>0.25].index].corr(), annot=True, fmt='.2f', cmap='Reds')
plt.title('Correlation Matrix of Highly Correlated Features')
plt.show()
```

```python
df = df.drop(columns=['GarageYrBlt', 'GarageCars','TotRmsAbvGrd','Foundation','2ndFlrSF','HalfBath','TotalBsmtSF'])
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean()
                                             )
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns='SalePrice' , axis=1)
y = df['SalePrice']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x
```

```python
y
```

```python
model = XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.1,
    max_depth=6,
    n_estimators=1000,
    subsample=0.65,
    random_state=42
)

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```

```python
import joblib

joblib.dump(model, "xgboost_model.pkl")

```
